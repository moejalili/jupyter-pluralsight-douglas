__we can access the style property of the data frame by running this below function__

```python
df_bus.style.highlight_max()
```

```python
def highlight_10000_and_over(number):
    if isinstance(number, int):
        return 'color: {0}'.format('green' if number >= 100000 else 'black')
    return 'color: black'
```

```python
df_bus.style.applymap(highlight_10000_and_over)     #Doesn't show this line ????
```
